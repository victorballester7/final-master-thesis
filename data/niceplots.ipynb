```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read_data(file_path):
    # Placeholder for the actual implementation of read_data
    return np.loadtxt(file_path)
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

problem = 'spread_2DNS'
kdn = '03-kdn32'
quantity = 'Energy'
# quantity = 'Enstrophy'
tests = ['test5', 'test6', 'test7']
labels = ['Re = 8', 'Re = 16', 'Re = 32']

fig, ax = plt.subplots()

data = []
for test in tests:
    data.append(read_data('./' + problem + '/' + kdn + '/' + test + '/' + quantity + 'MeanRadius.txt'))



domain = 4096

for i in range(len(data)):
    data[i][:,1] *= (2 * np.pi / domain)
    ax.plot(data[i][:,0], data[i][:,1], label=labels[i])

ax.legend()

# decorations
ax.set_xlabel('Time')
ax.set_ylabel('Mean radius')

ax.set_xlim(0,2.6)

plt.savefig('../images/' + problem + '/' + kdn + '/' + quantity + 'MeanRadius.pdf', bbox_inches='tight')
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family='serif')

problem = 'embarrassed_2DNS'
kdn = '02-kdn16'
quantity = 'Energy'
times = ['040', '080', '130', '179']
# quantity = 'Enstrophy'
test = 'test6'
labels = ['$t = 0.4$', '$t = 0.8$', '$t = 1.3$', '$t = 1.79$']

fig, ax = plt.subplots()

data = []
for time in times:
    data.append(
        read_data(
            './' +
            problem +
            '/' +
            kdn +
            '/' +
            test +
            '/' +
            quantity + '.' + time + '.txt'))

domain = 2048



for i in range(len(data)):
    data[i][:, 0] = np.arange(0, len(data[i][:, 0])) + 1
    data[i][:, 0] *= (2 * np.pi / domain)
    ax.plot(data[i][:, 0], data[i][:, 1], label=labels[i])

ax.legend()

# decorations
ax.set_xlabel('Radius')
ax.set_ylabel('Energy')

# ax.set_xlim(0, 2.)

plt.savefig(
    '../images/' +
    problem +
    '/' +
    kdn +
    '/' +
    quantity +
    'Prof_' + test + '.pdf', 
    bbox_inches='tight')
```
